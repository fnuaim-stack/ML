# Lab 3: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv("instagram_usage_lifestyle.csv")
df.head()
```

### Check Missing Values

```python
print(df.isna())
```

```python
print(df.isna().sum())
```

### Check duplicate rows

```python
df.duplicated()[df.duplicated()==True]
```

### No. of rows and columns

```python

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

### Data type of columns

```python
df.dtypes
```

```python
df['last_login_date'] = pd.to_datetime(df['last_login_date'])

df.dtypes
```

### Descriptive summary Statistics

```python
df.describe(include='all')
```

### Univariate Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(df['sleep_hours_per_night'], bins=20)
plt.title("Distribution of Sleep Hours Per Night")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()
```

The histogram represents the distribution of sleep hours per night, indicating the most common sleep durations and variability within the dataset.

### Distribution of User Engagement Score

```python
plt.figure(figsize=(8,5))
sns.histplot(df['user_engagement_score'], bins=20)
plt.title("Distribution of User Engagement Score")
plt.xlabel("Engagement Score")
plt.ylabel("Frequency")
plt.show()
```

## Bivariate Analysis

### Instagram Usage by Diet Quality

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='diet_quality',
            y='daily_active_minutes_instagram',
            data=df)

plt.title("Instagram Usage vs Diet Quality")
plt.show()
```

### Relationship Between Instagram Usage and Engagement Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='daily_active_minutes_instagram',
                y='user_engagement_score',
                data=df)

plt.title("Instagram Usage vs Engagement Score")
plt.xlabel("Daily Active Minutes on Instagram")
plt.ylabel("User Engagement Score")
plt.show()
```

### Correlation Matrix

```python
corr = df.select_dtypes(include='number').corr()

# Keep only strong correlations
filtered_corr = corr[(corr > 0.4) | (corr < -0.4)]

plt.figure(figsize=(12,10))
plt.imshow(filtered_corr, aspect='auto')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Strong Correlations Between Numeric Features")
plt.tight_layout()

plt.show()
```

- Correlation close to 1 = strong positive relationship
- Close to 0 = weak relationship

## Time-Based Analysis
### User Activity Over Time

```python
df['Month'] = df['last_login_date'].dt.to_period('M')

monthly_activity = df.groupby('Month').size()

plt.figure(figsize=(10,5))
monthly_activity.plot()

plt.title("User Activity Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Logins")
plt.show()
```
